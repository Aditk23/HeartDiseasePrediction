```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# column names
columns=['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','num']

# clevaland 
df_cleveland=pd.read_csv('processed.cleveland.data',header=None,names=columns)
df_cleveland.NAME_='cleveland'
# hungarian 
df_hungarian=pd.read_csv('processed.hungarian.data',header=None,names=columns)
df_hungarian.NAME_='hungarian'
# switzerland 
df_switzerland=pd.read_csv('processed.switzerland.data',header=None,names=columns)
df_switzerland.NAME_='switzerland'
# va 
df_va=pd.read_csv('processed.va.data',header=None,names=columns)
df_va.NAME_='va'

# see shape of all dataframes 
['Name:{} - Row: {}, columns:{}'.format(df_.NAME_,*df_.shape) for df_ in [df_cleveland,df_hungarian,df_switzerland,df_va]]
```

```python
# Adding column city to distinguish all 4 datasets and concatenating them

df_cleveland['city'] = 'Clevelend'
df_hungarian['city'] = 'Hungary'
df_switzerland['city'] = 'Switzerland'
df_va['city'] = 'VA'

all_data = pd.concat([df_cleveland,df_hungarian,df_switzerland,df_va])
all_data.head()
```

```python
# In this dataset null values are indicated by ? hence this method gives 0 null values
all_data.isnull().sum()
```

```python
# column wise count of null values
for column_ in all_data.columns:
  print(f'{column_}:',list(all_data[column_].values.reshape(-1)).count('?'))
```

```python
# replace '?' with 'NaN'
all_data.replace(to_replace='?',value=np.NaN,inplace=True)

# re-check for '?'
list(all_data.where(all_data=='?').fillna(0).values.reshape(-1)).count('?')
```

```python
# Both ca and thal are categorical and are having almost all missing values for Hungary and Switzerland
# But these all features are important as per study so connot drop any
c1 = all_data.loc[(all_data.city=='Switzerland')]
c2 = all_data.loc[(all_data.city=='Hungary')]
c3 = all_data.loc[(all_data.city=='VA')]
c4 = all_data.loc[(all_data.city=='Clevelend')]

```

```python
c1['ca'] = c1['ca'].fillna(c1['ca'].mode()[0])
c2['ca'] = c2['ca'].fillna(c2['ca'].mode()[0])
c3['ca'] = c3['ca'].fillna(c3['ca'].mode()[0])

c1['thal'] = c1['thal'].fillna(c1['thal'].mode()[0])
c2['thal'] = c2['thal'].fillna(c2['thal'].mode()[0])
c3['thal'] = c3['thal'].fillna(c3['thal'].mode()[0])

```

```python
all_data_new = pd.concat([c4,c1,c2,c3])
```

```python
all_data_new = all_data_new.ffill(axis = 0)

```
